```python
import pandas as pd
import re
import os
import shutil
from dateutil.parser import parse
```

```python
def read_txtrisc(filename):
    try:
        content = open(filename,'r').read().lower()
        startpos = content.find('survey check'.lower())
        if startpos<0: startpos=0
        if startpos>500: startpos=500
        endpos = content.find('Submitted by SCC'.lower())
        return content[startpos:endpos]
    except:
        pass
def empty_folder(folder):
    for filename in os.listdir(folder):
        file_path = os.path.join(folder, filename)
        try:
            if os.path.isfile(file_path) or os.path.islink(file_path):
                os.unlink(file_path)
            elif os.path.isdir(file_path):
                shutil.rmtree(file_path)
        except Exception as e:
            print('Failed to delete %s. Reason: %s' % (file_path, e))
def showsearched(search_df):
    search_output_folder   = 'D:/tmrisc/searchs/'
    if search_df['txt'].count()>0:
        print('found {} files'.format( search_df['request_no'].count()))
        empty_folder(search_output_folder)
        for jpg_filename in search_df['jpg']:
            shutil.copy(jpg_filename,search_output_folder)
    else:
        print('found none')
def moresearch(search_df,patternstring):
    output = search_df['txt'].map(read_txtrisc)
    return search_df[output.str.contains(patternstring)]
```

```python
df = pd.read_csv('d:/tmrisc/risc_data.csv')
df.columns
```

```python
#Seach by Station
search_df = df[df['thaodien']*df['ceiling']]
search_df.txt.count()
```

```python
#More Search
search_df = moresearch(search_df,'final')
search_df.txt.count()
```
